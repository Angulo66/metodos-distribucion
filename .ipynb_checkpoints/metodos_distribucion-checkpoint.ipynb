# Distribucion Binomial

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
n = 10
p = 0.3
k = np.arange(0,39)
binomial = stats.binom.pmf(k, n, p)
binomial
```

```python
plt.plot(k, binomial, 'o-')
plt.title('Binomial: n=%i, p=%.2f' % (n,p), fontsize=15)
plt.xlabel('Number of successes')
plt.ylabel('Probability of successes', fontsize=15)
plt.show()
```

# Distribucion Poisson

```python
rate = 36 
n = np.arange(0, 39)
y = stats.poisson.pmf(n, rate)
y
```

```python
plt.plot(n, y, 'o-')
plt.title("Poisson: $\lambda$ =%i" % rate)
plt.xlabel('Number of accidents')
plt.ylabel('Probability of number of accidents')
plt.show()
```

```python
data = stats.poisson.rvs(mu=2, loc=0, size=1000)
print("Mean: %g" % np.mean(data))
print("SD: %g" % np.std(data,ddof=1))
plt.figure()
plt.hist(data, bins=9, density=True)
plt.xlim(0,10)
plt.xlabel("Number of accidents")
plt.title("Simulating Poisson Random Variables")
plt.show()
```

# Distribucion Normal 

```python
mu = 0 # mean
sigma = 1 # standard deviation
x = np.arange(-5,5,0.1)
y = stats.norm.pdf(x, 0, 1)
plt.plot(x, y)
plt.title('Normal: $\mu$=%.1f, $\sigma^2$=%.1f' % (mu, sigma))
plt.xlabel('x')
plt.ylabel('Probability density') # probability of observing each of these observations
plt.show()
```
